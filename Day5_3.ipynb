```python
import tensorflow as tf
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
crab = pd.read_csv(r"C:\Users\user\Desktop\bigdata_8\data\CrabAgePrediction.csv")
crab
```

```python
input = crab.drop(columns=['Age','Sex']).values
target = crab['Age'].values
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(input,
                                                                      target,
                                                                      test_size=0.2,
                                                                      random_state=27)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, input_dim = 7, activation = 'relu'),
    tf.keras.layers.Dense(30, activation = 'relu'),
    tf.keras.layers.Dense(1)
]) 
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mean_squared_error'])
```

```python
# 정규화 작업을 진행합니다.
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_input)
test_scaled = scaler.transform(test_input)
```

```python
model.fit(train_scaled, train_target, epochs = 20 , batch_size = 10)
```

```python
model.evaluate(train_scaled, train_target)
```

```python
model.evaluate(test_scaled, test_target)
```

```python
# 예측
predictions = model.predict(test_scaled)
print("예측 결과:", predictions)
```
